```python
import pandas as pd  

# Upload the dataset or load it from a URL
df = pd.read_csv("most_watched_anime_dataset_100_entries.csv")

# Display basic info
df.head()  # Show first 5 rows
df.info()  # Summary of dataset
df.describe()  # Summary statistics (numerical columns)
df.isnull().sum()  # Check for missing values

```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Univariate Analysis
for col in df.select_dtypes(include=['object']).columns:
# Load the dataset
    print(col, ' : ', df[col].nunique())
    print(df[col].value_counts(normalize=True).head())
```

```python

# Bivariate Analysis: Scatter plot of 'Ratings' vs 'Number of Episodes'
plt.figure(figsize=(10, 6))
plt.scatter(df['Number of Episodes'], df['Ratings'], alpha=0.5)
plt.title('Scatter Plot of Ratings vs Number of Episodes')
plt.xlabel('Number of Episodes')
plt.ylabel('Ratings')
plt.grid(True)
plt.show()

```

```python
import pandas as pd
# 1. Correlation Heatmap
plt.figure(figsize=(12, 8))
numerical_features = ['Number of Episodes', 'Budget (in Million USD)', 'Ratings']
df1 = df[numerical_features].apply(pd.to_numeric, errors='coerce').dropna()
# Load the dataset
sns.heatmap(df1.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")

# Upload the dataset or load it from a URL
plt.show()
```

```python
# 2. Pairplot (Scatterplot Matrix)
sns.pairplot(df1, diag_kind="kde", corner=True)
plt.show()
# 3. Scatter Matrix (Alternative to Pairplot)
scatter_matrix(df1, figsize=(12, 8), diagonal='kde', alpha=0.6)
plt.show()
```

```python
# 5. Boxplot: Track Popularity by Playlist Genre
plt.figure(figsize=(12,6))
sns.boxplot(x=df["Number of Episodes"], y=df["Budget (in Million USD)"], palette="Set2")
plt.xticks(rotation=90)  # Rotate labels for better visibility
plt.title("Boxplot of Number of Episodes vs Budget (in Million USD)")
plt.xlabel("Number of Episodes")
plt.ylabel("Budget (in Million USD)")
plt.show()
plt.figure(figsize=(12, 6))

```

```python
#6 Histogram
sns.histplot(data=df, x="Number of Episodes", hue="Ratings",multiple="stack", bins=30, palette="Set2")
plt.title("Histogram of Number of Episodes by Ratings")
plt.xlabel("Number of Episodes")
plt.ylabel("Ratings")
plt.legend(title="Ratings", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
